# One Hot Encoding and Dummy Data

```python
import pandas as pd 

df = pd.read_csv('homeprices.csv')
df
```

### Dummydata with pandas function

```python
dummydf=pd.get_dummies(df.town)
dummydf
```

```python
merged_df = pd.concat([df, dummydf], axis='columns')
merged_df
```

```python
cleandf = merged_df.drop(['town', 'monroe township'], axis=1)
cleandf
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(cleandf[['area', 'robinsville', 'west windsor']].values, df.price)
```

```python
model.predict([[2800, 1, 0]])
```

```python
model.predict([[3400, 0, 1]])
```

```python
model.score(cleandf[['area', 'robinsville', 'west windsor']].values, df.price)
```

### One hot encoding scikit

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town', 'area']].values
X
```

```python
y= dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.predict([[1, 0, 2800]])
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(cleandf[['Mileage', 'Age', 'BMW X5', 'Mercedez Benz C class']], df.Sell_Price)
```

```python
model.predict([[45000, 4, 0, 1]])
```

```python
model.predict([[86000, 7, 1, 0]])
```

```python
model.score(cleandf[['Mileage', 'Age', 'BMW X5', 'Mercedez Benz C class']], df.Sell_Price)
```
